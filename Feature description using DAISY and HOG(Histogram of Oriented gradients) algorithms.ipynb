```python
import matplotlib.pyplot as plt
from skimage.feature import daisy
from skimage import color,data,io
import skimage
```

```python
mickey_image = skimage.io.imread("mickey-mouse-girl-friend-image-desktop.jpg")
mickey_image=color.rgb2gray(mickey_image)
plt.imshow(mickey_image,cmap='gray')
```

```python
descp_array,descp_image=daisy(mickey_image,step=200,radius=20,visualize=True)
```

```python
plt.imshow(descp_image)
```

# HOG

```python
from skimage.feature import hog
from skimage import data,io,color,exposure
```

```python
fd,hog_image = hog(mickey_image,
                  pixels_per_cell=(10,10),
                  block_norm = 'L2-Hys',
                  visualize=True,
                  multichannel=False)
```

```python
plt.imshow(hog_image)
```

# To get higher contrast image

```python
hog_iamge_rescaled= exposure.rescale_intensity(hog_image,in_range=(0,10))
```

```python
plt.imshow(hog_iamge_rescaled,cmap="gray")
```
