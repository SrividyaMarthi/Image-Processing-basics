```python
import math
import matplotlib.pyplot as plt
import skimage
from skimage import io
from skimage import data
from skimage import transform 
from skimage.transform import swirl,warp
```

```python
mickey_image =skimage.io.imread("mickey-mouse-girl-friend-image-desktop.jpg")
plt.imshow(mickey_image)
```

```python
tform  = transform.SimilarityTransform(scale = 1.5,  rotation=math.pi/4,
                                      translation=(mickey_image.shape[0]/2,-100))
```

```python
rotated = warp(mickey_image,tform)
back_rotated = warp(rotated, tform.inverse)
```

```python
plt.imshow(rotated)
```

```python
plt.imshow(back_rotated)
```

```python
mickey_swirled = swirl(mickey_image,rotation=0, strength=10,radius=330,mode='constant')
```

```python
plt.imshow(mickey_swirled)
```
