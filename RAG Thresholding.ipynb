```python
from matplotlib import pyplot as plt
import skimage
from skimage import data,io,segmentation,color
from skimage.future import graph
from skimage import io
```

```python
mickey_image =skimage.io.imread("mickey-mouse-girl-friend-image-desktop.jpg")
plt.imshow(mickey_image)
```

```python
# segmenattaion used k-means clustering
lables_l = segmentation.slic(mickey_image,compactness=35,n_segments=500)
```

```python
lables_l.shape
```

```python
lables_l
```

```python
segmented_overlay = color.label2rgb(lables_l,mickey_image,kind="avg")

plt.figure(figsize=(8,8))
plt.imshow(segmented_overlay)
```

```python
g = graph.rag_mean_color(mickey_image,lables_l)
```

```python
labels_2 = graph.cut_threshold(lables_l,g,thresh=15)
```

```python
segmented_rag = color.label2rgb(labels_2,mickey_image,kind='avg')

plt.figure(figsize=(8,8))
plt.imshow(segmented_rag)
```
