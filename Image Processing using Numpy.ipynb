# Block views and Pooling Operations

```python
import skimage.io
from skimage import color
from skimage.util import view_as_blocks
```

```python
block_view_mickey_image=color.rgb2gray(mickey_image)
```

```python
plt.imshow(block_view_mickey_image,cmap="gray")
```

```python
block_view_mickey_image.shape
```

```python
block_shape=(4,4)
```

```python
Mickey_blocks=view_as_blocks(block_view_mickey_image,block_shape)
```

```python
Mickey_blocks.shape
```

```python
flattened_blocks = Mickey_blocks.reshape(Mickey_blocks.shape[0],Mickey_blocks.shape[1],-1)
```

```python
print("Shape of original image:",Mickey_blocks.shape)
print("Shape of flattened image:",flattened_blocks.shape)
```

# Pooling Aggregations

```python
# To find the mean value of each block
mean_blocks =np.mean(flattened_blocks,axis=2)
plt.imshow(mean_blocks,interpolation='nearest',cmap='gray')
```

```python
# To find the max value of each block
Max_blocks =np.max(flattened_blocks,axis=2)
plt.imshow(Max_blocks,interpolation='nearest',cmap='gray')
```

```python
# To find the median value of each block
Median_blocks =np.median(flattened_blocks,axis=2)
plt.imshow(Median_blocks,interpolation='nearest',cmap='gray')
```
