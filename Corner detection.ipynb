```python
from matplotlib import pyplot as plt
from skimage import data
from skimage.feature import corner_harris, corner_subpix, corner_peaks
from skimage.transform import warp, AffineTransform
from skimage.draw import ellipse,circle
```

```python
checkboard = data.checkerboard()
```

```python
plt.imshow(checkboard,cmap="gray")
```

```python
transform = AffineTransform(scale=(0.9,0.8),rotation=1,shear=0.6,translation=(150,-80))
```

```python
warped_checkerboard = warp(checkboard,transform,output_shape=(320,320))
```

```python
plt.imshow(warped_checkerboard,cmap="gray")
```

```python
corners = corner_harris(warped_checkerboard)
plt.imshow(corners,cmap="gray")
```

```python
corner_coordinates = corner_peaks(corners,min_distance=1)
corner_coordinates.shape
```

```python
# statistical dettection of corners which are more accurate comparativey
coords_subpix = corner_subpix(warped_checkerboard, corner_coordinates, window_size=10)
```
