```python
import numpy as np
import matplotlib.pyplot as plt
import skimage
from skimage import io
from skimage import data, img_as_float,color
from skimage.measure import compare_ssim as ssim
```

```python
mickey_image =skimage.io.imread("mickey-mouse-girl-friend-image-desktop.jpg")
plt.imshow(mickey_image)
```

```python
noise = np.ones_like(mickey_image)*0.4*(mickey_image.max()-mickey_image.min())
```

```python
noise[np.random.random(size=noise.shape)>0.7]*=-1
```

```python
# MSE calculates the mean square error between two input images
def mse(x,y):
    return np.linalg.norm(x-y)
```

```python
noisy_mickey = mickey_image + noise
plt.imshow(noisy_mickey)
```

```python
mickey_noisy_constant = mickey_image +abs(noise)
plt.imshow(mickey_noisy_constant)
```

```python
# comparing original
mse_none =mse(mickey_image,mickey_image)

seim_moise = ssim(mickey_image,mickey_image,
                 multichannel=True,
                 data_range=mickey_image.max() - mickey_image.min())
```

```python
print(mse_none,",",seim_moise)
```

```python
# comparing noise 1
mse_noise1 =mse(mickey_image,noisy_mickey)

seim_moise1 = ssim(mickey_image,noisy_mickey,
                 multichannel=True,
                 data_range=mickey_image.max() - mickey_image.min())
```

```python
print(mse_noise1,",",seim_moise1)
```

```python
# comparing noise 2
mse_noise2=mse(mickey_image,mickey_noisy_constant)

seim_moise2 = ssim(mickey_image,mickey_noisy_constant,
                 multichannel=True,
                 data_range=mickey_image.max() - mickey_image.min())
```

```python
print(mse_noise2,",",seim_moise2)
```
