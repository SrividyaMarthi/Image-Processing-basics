```python
import skimage.io
from skimage import color
from skimage.util import view_as_blocks

from skimage import measure
import numpy
import matplotlib.pyplot as plt
```

```python
mickey_image=skimage.io.imread("mickey-mouse-girl-friend-image-desktop.jpg")
```

```python
mickey_image.shape
```

```python
color_image = color.rgb2gray(mickey_image)
plt.imshow(color_image,cmap="gray")
```

```python
contours = measure.find_contours(color_image,0.5)
len(contours)
```

```python
fig ,axes = plt.subplots(1,2,figsize=(16,12),sharex=True,sharey=True)

ax=axes.ravel()
ax[0].imshow(color_image,cmap="gray")
ax[0].set_title("Original image",fontsize=20)

ax[1].imshow(color_image,interpolation="nearest",cmap="gray")
ax[1].set_title("Contour image",fontsize=20)

for n,contour in enumerate(contours):
    ax[1].plot(contour[:,1],contour[:,0],linewidth=5)
    
plt.show()
```

# Generating Contours

```python
import numpy as np

A,B = np.ogrid[-np.pi:np.pi:100j, -np.pi:np.pi:100j]
complex_image=np.sin(np.exp((np.sin(A)**3 + np.cos(B)**3)))
complex_image.shape
```

```python
plt.imshow(complex_image,cmap="gray")
```

```python
contours = measure.find_contours(complex_image,0.5)
len(contours)
```

```python
fig ,axes = plt.subplots(1,2,figsize=(16,12),sharex=True,sharey=True)

ax=axes.ravel()
ax[0].imshow(complex_image,cmap="gray")
ax[0].set_title("Original image",fontsize=20)

ax[1].imshow(complex_image,interpolation="nearest",cmap="gray")
ax[1].set_title("Contour image",fontsize=20)

for n,contour in enumerate(contours):
    ax[1].plot(contour[:,1],contour[:,0],linewidth=5)
    
plt.show()
```
