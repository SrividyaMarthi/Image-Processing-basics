```python
import matplotlib.pyplot as plt
import skimage
from skimage import data,io,color
from skimage.filters import threshold_otsu,threshold_local
```

```python
mickey_image =color.rgb2gray(skimage.io.imread("mickey-mouse-girl-friend-image-desktop.jpg"))
plt.imshow(mickey_image,cmap='gray')
```

```python
mickey_image= mickey_image*255
thres= threshold_otsu(mickey_image)
print(thres)
```

```python
mickey_binary = mickey_image > thres
```

```python
mickey_binary
```

```python
plt.imshow(mickey_binary,cmap='gray')
```

```python
fig,ax = plt.subplots(figsize=(6,6))

ax.hist(mickey_image.ravel())
ax.set_title('Histogram')
ax.set_xlabel('Pixel Intensity')
ax.set_ylabel('Pixel Count')

ax.axvline(thres,color='y')
```

```python
# To get the best intensity
from skimage.filters import try_all_threshold

fig,ax = try_all_threshold(mickey_image,figsize=(6,12),verbose=False)
plt.show()
```
