# Used in Applications :Pattern recognition,Statistics and in image processing

```python
import skimage.io
from skimage import data,img_as_float
from skimage.morphology import convex_hull_image

from skimage.util import invert
import numpy as np
import matplotlib.pyplot as plt
```

```python
horse_image=data.horse()
```

```python
inverted_horse=invert(horse_image)
```

```python
fig ,axes = plt.subplots(1,2,figsize=(16,12),sharex=True,sharey=True)

ax=axes.ravel()
ax[0].imshow(horse_image,cmap="gray")
ax[0].set_title("Original image",fontsize=20)

ax[1].imshow(inverted_horse,interpolation="nearest",cmap="gray")
ax[1].set_title("Inverted image",fontsize=20)


fig.tight_layout()
plt.show()
```

```python
convexhull_image = convex_hull_image(inverted_horse)
```

```python
fig ,axes = plt.subplots(1,2,figsize=(16,12),sharex=True,sharey=True)

ax=axes.ravel()
ax[0].imshow(inverted_horse,cmap="gray")
ax[0].set_title("inverted horse image",fontsize=20)

ax[1].imshow(convexhull_image,interpolation="nearest",cmap="gray")
ax[1].set_title("convexhull image",fontsize=20)


fig.tight_layout()
plt.show()
```

```python
convexhull_diff = img_as_float(convexhull_image.copy())
```

```python
convexhull_diff[inverted_horse]=2
```

```python
fig ,ax = plt.subplots(figsize=(16,12))

ax.imshow(convexhull_diff,cmap="gray")
ax.set_title("convexhull image",fontsize=20)
    
plt.show()
```
